```
from pyspark.sql.functions import col
#Import to use the date_format function
from pyspark.sql import functions as F
from pyspark.sql.functions import weekofyear, hour, month

service_credential = 

spark.conf.set("fs.azure.account.auth.type.ade20220919.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.ade20220919.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.ade20220919.dfs.core.windows.net", "98a69314-1a2f-4eef-9b09-6ff1e73ead68")
spark.conf.set("fs.azure.account.oauth2.client.secret.ade20220919.dfs.core.windows.net", service_credential)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.ade20220919.dfs.core.windows.net", "https://login.microsoftonline.com/33da9f3f-4c1a-4640-8ce1-3f63024aea1d/oauth2/token")

df_main_event = spark.read.parquet(f"abfs://silver-layer@ade20220919.dfs.core.windows.net/K-B-B/gharchive/event_table/day=*/*.snappy.parquet")
df_push_event = spark.read.parquet(f"abfs://silver-layer@ade20220919.dfs.core.windows.net/K-B-B/gharchive/push_event_table/day=*/*.snappy.parquet")
df_member_event = spark.read.parquet(f"abfs://silver-layer@ade20220919.dfs.core.windows.net/K-B-B/gharchive/member_event_table/day=*/*.snappy.parquet")
```

```
#User activity should be aggregated so that a filterable chart can be populated with breakdowns of user activity by week or month.

#Added hour, day, and month columns extracted from the "created-at" column, which can then be used for filtering in the future

df_main_event = df_main_event.withColumn('week_of_year', weekofyear(df_main_event.created_at))\
    .withColumn('hour', hour(df_main_event.created_at)) \
    .withColumn("day", F.date_format("created_at", "dd")) \
    .withColumn('month', month(df_main_event.created_at))

display(df_main_event)
```

```
#Data aggregated by type of GitHub event per hour

#Grouped event records by hour and type and counted the totals of each

df_hourly_agg = df_main_event.groupBy('hour', 'type')\
    .count()\
    .sort('hour', 'count', ascending=False)

display(df_hourly_agg)
```

```
#PushEvent data aggregated by ref  whether the commit is on the main branch

#Grouped push events by ref type, and then counted up each row that was aggregated

df_ref_agg = df_push_event.groupBy('ref').count().sort('count', ascending=False)

display(df_ref_agg)

```

```
#Breakdown of number of commits per PushEvent

#Commits/Rows grouped by push_id, so single push for multiple commits were counted. 

df_sum_of_commits = df_push_event.groupBy('push_id').agg({'distinct_size' : 'sum'})\
    .sort('sum(distinct_size)', ascending=False)\
    .withColumnRenamed('sum(distinct_size)', 'total_commits')

display(df_sum_of_commits)
    
```

```
df_sum_of_commits.describe('total_commits').show()
```

```
# Use case: Show the repo with the most members added
df_exp = df_main_event.join(df_member_event, 'id')

df_exp_added = df_exp.where(col('action') == 'added') \
    .groupBy('name').count()\
    .sort('count', ascending=False)

display(df_exp_added)
```

```
#What are the most common actions per repo?
df_repo_usage = df_main_event.groupBy('repo_id', 'type').count()\
    .sort('count', ascending=False)\

display(df_repo_usage)
```
